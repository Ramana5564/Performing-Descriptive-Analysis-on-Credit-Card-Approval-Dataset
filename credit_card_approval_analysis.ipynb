**1.List the column names which hold numerical data, and which hold categorical data, separately.**

---



```python
column_names = credit_card_data.columns
print(column_names)
```

* **Now we should seperate the numerical and categorical columns**




```python
numerical_columns = credit_card_data.select_dtypes(include=['int', 'float', 'double']).columns
categorical_columns = credit_card_data.select_dtypes(include=['object', 'category', 'bool']).columns
```

```python
# print numerical columns data
print("Numerical columns :",numerical_columns.tolist())
```

```python
# print categorical columns data
print("Categorical columns : ",categorical_columns.tolist())
```

**2.Check if there are any duplicate rows in the data. If present, drop them**
---
---



* **Using Pandas library to check the duplicate rows in the dataset**

```python
# "duplicated()" is a function in pandas library which checks duplicate rows in dataset
duplicate_rows = credit_card_data.duplicated().sum()
if duplicate_rows > 0:
    print(duplicate_rows)
    credit_card_data.drop_duplicates(inplace=True)
    print("Duplicate rows have been removed.")
else:
    print("No duplicate rows found.")
```

**4. List the descriptive statistics for each numeric column**
---
---

```python
descriptive_statistics = credit_card_data.describe()
print("Descriptive Statistics of Credit Card Approval Dataset \n",descriptive_statistics)
```

**5. Plot histogram for each numeric column.**
---
---

```python
credit_card_data[numerical_columns].hist(figsize=(15, 15))
plt.show()
```

**6. Plot count plot for each categorical column.**
---
---

```python
categorical_columns
```

```python
# Set up subplots based on the number of categorical columns
num_columns = len(categorical_columns)
fig, axes = plt.subplots(num_columns, 1, figsize=(10, 6 * num_columns))
plt.subplots_adjust(hspace=0.5)  # Adjust vertical spacing between subplots

# Create count plots for each categorical column
for i, col in enumerate(categorical_columns):
    sns.countplot(data=credit_card_data, x=col, ax=axes[i])
    axes[i].set_title(f'Count Plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=40)
    print("")

plt.show()
```

**7. Plot pairplot for all numeric features.**
---
---

```python
sns.pairplot(credit_card_data[numerical_columns], hue='Approved', diag_kind='kde')
plt.show()
```

**8. Check if the dataset is balanced or imbalanced.**
---
---

```python
# Count the occurrences of each class
class_counts = credit_card_data["Approved"].value_counts()

# Visualize the class distribution
plt.figure(figsize=(6, 4))
class_counts.plot(kind='bar', color='skyblue')
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()

# Calculate class ratios
total_samples = len(credit_card_data)
class_ratios = class_counts / total_samples

print("Class Ratios:")
print(class_ratios)
```

* **If the ratios are roughly equal, the dataset is balanced. If some ratios are much smaller, it's imbalanced.**
* **Here ratios are almost equal so we can say that the dataset is balanced.**

**9. Plot Heat Map with Correlation Coefficient for all numeric features.**
---
---


```python
coefficient_matrix = credit_card_data[numerical_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(coefficient_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

**10. Plot Kernel Density Estimate plot for 'Age', 'CreditScore' and 'YearsEmployed' with target feature 'Approved' as Hue**
---
---

```python
# Subset the dataset to include only the relevant columns
subset_data = credit_card_data[['Age', 'CreditScore', 'YearsEmployed', 'Approved']]
# Create KDE plots with 'Approved' as Hue
plt.figure(figsize=(12, 6))
for feature in ['Age', 'CreditScore', 'YearsEmployed']:
    sns.kdeplot(data=subset_data, x=feature, hue='Approved', fill=True, common_norm=False, alpha=0.5, label=feature)
plt.title("Kernel Density Estimate (KDE) Plots")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend(title='Approved', loc='best')
plt.show()
```

**11. Generate cross table for 'Gender' and 'Approved' columns.**
---
---

```python
cross_table = pd.crosstab(credit_card_data['Gender'], credit_card_data['Approved'])
print("\nCross Table for 'Gender' and 'Approved' columns:")
print(cross_table)
```

**12. Generate bar plots for â€˜Ethnicityâ€™, â€˜Employedâ€™ and â€˜BankCustomerâ€™ columns with target feature â€˜Approvedâ€™ as Hue**
---
---

```python
# Create bar plots

Set1 = ['skyblue', 'orange']

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.countplot(data=credit_card_data, x='Ethnicity', hue='Approved')
plt.title('Ethnicity vs. Approved')

plt.subplot(1, 3, 2)
sns.countplot(data=credit_card_data, x='Employed', hue='Approved', palette='Set2')
plt.title('Employed vs. Approved')

plt.subplot(1, 3, 3)
sns.countplot(data=credit_card_data, x='BankCustomer', hue='Approved', palette=Set1)
plt.title('BankCustomer vs. Approved')

plt.tight_layout()
plt.show()
```

**13. Generate FacetGrid with â€˜Marriedâ€™ as column, â€˜Employedâ€™ as row, â€˜Ageâ€™ as X axis, â€˜YearsEmployedâ€™ as Y axis and â€˜Approvedâ€™ as Hue.**
---
---

```python
# Create the FacetGrid
ramana = sns.FacetGrid(credit_card_data, col='Married', row='Employed', hue='Approved', margin_titles=True)

# Define the scatter plot using the 'Age' and 'YearsEmployed' columns
ramana.map(plt.scatter, 'Age', 'YearsEmployed')

# Add legend
ramana.add_legend()

# Set titles
ramana.set_axis_labels("Age", "YearsEmployed")

# Set overall title
ramana  .fig.suptitle("FacetGrid: Married vs. Employed vs. Age vs. YearsEmployed vs. Approved", fontsize=16)

# Adjust the layout
plt.subplots_adjust(top=0.9)

# Show the plot
plt.show()
```
